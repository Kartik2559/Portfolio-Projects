```python
# Reading the CSV Data using Pandas
```

```python
df = pd.read_csv('/Users/kartik/Downloads/movies.csv')
```

```python
df.head(10) #Looking at top 10 rows of Data just for reference
```

```python
# Looking if there is any missing data or not
```

```python
for col in df.columns:
    Null_Percentage = np.mean(df[col].isna())
    print(f'{col} - {Null_Percentage}%')
```

```python
df = df.fillna(0) #Replacing all the Null values in data with '0'
```

```python
# Compare %age value of Null's in the Data

for col in df.columns:
    Null_Percentage = np.mean(df[col].isna())
    print(f'{col} - {Null_Percentage}%')
```

```python
df.dtypes #Checking Data Types of columnes in Data
```

```python
# Converting all unnecessary Float values to intiger values
for col in df.columns:
    if col == 'score':
        pass
    elif df[col].dtype == 'float64':
        df[col] = df[col].astype(int)
```

```python
# Creating CorrectYear column referning released
df['Correct_Year'] = df['released'].astype(str).str.split(',').str[1].str[1:5]
```

```python
df = df.fillna(0)
```

```python
# Arranging the data on the basis of Gross collected in decending order

df = df.sort_values(by = ['gross'], ascending = False)
```

```python
# Way to see entire data
# pd.set_option('display.max_rows', None)

pd.set_option('display.max_rows', 50) #Displaying first 50 rows of data
```

```python
df.drop_duplicates() #Drop duplicate data from entire table -- Just for self reference
```

```python

# There are two hypothesis that there is:

#     1. Correlation between Gross and Budget
#     2. Correlation between Company and Gross

# Have to perform tasks to come to conclusion whether hypothesis are correct or not.
```

```python
# Ploting Scatter Plot graph to understand the correlation between Gross earning and Budget of the Movie.
```

```python
plt.scatter(x = df['budget'], y = df['gross'] )
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Movie')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Creating a regression analysis graph
```

```python
sns.regplot(x ='budget', y = 'gross', data = df, scatter_kws = {'color':'#65598B'}, line_kws = {'color':'#73264C'})
plt.show()
```

```python
# Looking into the correlation by using default method = 'Pearson'
```

```python
df.corr() # differnt methods are 'pearson', 'kendall', 'spearman'. to access different methods we can use: df.corr(method = 'Name_of_method')
```

```python
# Creating correlation matrix heatmap using seaborn
```

```python
Correlation_Matrix = df.corr(method = 'pearson')
```

```python
sns.heatmap(Correlation_Matrix, annot=True)
plt.title('Correlation Matrix for Numeric Values')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df.head() # Showing first few rows to compare after assigning category codes
```

```python
# Converting all the object type data into category type and assigning them codes to create correlation matrix
df_numeric = df
for col in df_numeric.columns:
    if df_numeric[col].dtype == 'object':
        df_numeric[col] = df_numeric[col].astype('category')
        df_numeric[col] = df_numeric[col].cat.codes
```

```python
df_numeric.head() #Category Code table
```

```python
df_numeric_Correlation = df_numeric.corr() #Creating correlation of the all numeric data in table
```

```python
# Creating Heatmap for the correlation matrix

sns.heatmap(df_numeric_Correlation, annot = True)
plt.title('Table Correlation Heat Map')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
Correlation_Pairs = df_numeric_Correlation.unstack() #Creating Correlation pairs by using unstack() function
```

```python
Correlation_Pairs
```

```python
Sorted_Correlation_Pairs = Correlation_Pairs.sort_values(ascending = False) # Sorting correlation pairs in Decending Order
```

```python
pd.set_option('display.max_rows', None) # Displaying the entire Correlation Pairs
Sorted_Correlation_Pairs
```

```python
# Filtering the correlation values which are more than 0.5

High_Correlation = Sorted_Correlation_Pairs[0.5 <= Sorted_Correlation_Pairs ]

High_Correlation
```
